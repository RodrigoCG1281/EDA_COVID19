```python
##Librerias basicas a Importar
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

##LIMPIEZA DE DATOS
###Eliminar datos faltantes en algunas celdas



```python
umbral = 95315  # Umbral de registros mínimos

# Calcula el recuento de valores no nulos en cada columna
recuento_no_nulos = df.count()

# Obtiene el nombre de las columnas que tienen menos de 'umbral' registros no nulos
columnas_a_eliminar = recuento_no_nulos[recuento_no_nulos < umbral].index

# Elimina las columnas identificadas
df = df.drop(columnas_a_eliminar, axis=1)

# Muestra las dimensiones del DataFrame resultante
print("Dimensiones del DataFrame después de eliminar columnas: ", df.shape)

```

```python
df.info(verbose=True, show_counts=True)
```

```python
# Suponiendo que deseas exportar el DataFrame 'df' a un archivo llamado 'datos_limpios.xlsx'
nombre_archivo = 'datos_limpios.xlsx'

# Exporta el DataFrame a un archivo Excel
df.to_excel(nombre_archivo, index=False)  # El parámetro index=False evita que se incluya el índice en el archivo Excel

print(f"Los datos se han exportado correctamente a '{nombre_archivo}'.")
```

##GRAFICANDO

```python
# Contar la cantidad de personas contagiadas por género
cantidad_masculino = df[df['sexo'] == 'MASCULINO'].shape[0]
cantidad_femenino = df[df['sexo'] == 'FEMENINO'].shape[0]

# Crear una lista con los valores y etiquetas
valores = [cantidad_masculino, cantidad_femenino]
etiquetas = ['Masculino', 'Femenino']

# Crear la gráfica de barras
plt.bar(etiquetas, valores, color=['blue', 'pink'])

# Agregar título y etiquetas
plt.title('Cantidad de contagiados por género')
plt.xlabel('Género')
plt.ylabel('Cantidad de contagiados')

# Mostrar la gráfica
plt.show()
```

```python
# Definir las categorías de edad
categorias = ['Primera infancia', 'Infancia', 'Adolescencia', 'Juventud', 'Adultez', 'Vejez']
limites_edad = [0, 6, 12, 19, 27, 60, float('inf')]  # Definir los límites de edad para cada categoría

# Agregar una nueva columna con las categorías de edad
df['grupo_edad'] = pd.cut(df['edad'], bins=limites_edad, labels=categorias, right=False)

# Contar la cantidad de personas contagiadas en cada grupo de edad
conteo_grupo_edad = df['grupo_edad'].value_counts().sort_index()

# Mostrar el conteo
print(conteo_grupo_edad)

```

```python
# Conteo de personas contagiadas por grupo de edad
conteo_grupo_edad = df['grupo_edad'].value_counts().sort_index()

# Crear la gráfica
colores = ['blue', 'green', 'red', 'purple', 'orange', 'yellow']

plt.figure(figsize=(10, 6))
conteo_grupo_edad.plot(kind='bar', color=colores)

# Personalizar la gráfica
plt.title('Cantidad de personas contagiadas por grupo de edad')
plt.xlabel('Grupo de Edad')
plt.ylabel('Cantidad de Personas Contagiadas')
plt.xticks(rotation=45)

# Mostrar la gráfica
plt.show()
```

```python
# Contar la cantidad de personas en cada ocupación
conteo_ocupacion = df['ocupacion'].value_counts()

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
bars = plt.bar(conteo_ocupacion.index, conteo_ocupacion.values, color='lightblue')

# Agregar las etiquetas de texto con la cantidad exacta sobre cada barra
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), ha='center', va='bottom')

plt.title('Cantidad exacta de personas por ocupación')
plt.xlabel('Ocupación')
plt.ylabel('Cantidad de personas')
plt.xticks(rotation=45)  # Rotar las etiquetas del eje x para una mejor legibilidad
plt.tight_layout()  # Ajustar el diseño del gráfico
plt.show()
```

```python
direcciones_mayuscula = df['direccion'].str.upper()
conteo_por_hogar = direcciones_mayuscula.value_counts()
print(conteo_por_hogar)
```

```python
# Filtrar los lugares con menos de cierta cantidad de infectados (por ejemplo, 10)
lugares_filtrados = conteo_por_hogar[conteo_por_hogar <= 100 ].head(20)

# Graficar los lugares filtrados
lugares_filtrados.plot(kind='barh', color='skyblue', figsize=(10, 6))
plt.title('Lugares con Mayor Cantidad de Infectados (con menos de 100 casos)')
plt.xlabel('Cantidad de Infectados')
plt.ylabel('Lugar')
plt.show()



```
